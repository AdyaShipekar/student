```python
from IPython.display import HTML
import uuid, json

# --- Create a unique container ID to avoid conflicts on re-run ---
container_id = "timeline_" + str(uuid.uuid4()).replace("-", "_")

# --- Timeline data ---
timeline_data = [
    {
      "title": "Sprint 1: Tools & Equipment",
      "date": "Start of Trimester",
      "details": """Learned to use basic coding tools and troubleshoot issues.
                    <br><a href='https://adyashipekar.github.io/student/tools/blog' target='_blank'>View Guide</a>"""
    },
    {
      "title": "Sprint 2: Hacks & JavaScript Fundamentals",
      "date": "Mid Trimester",
      "details": """Built and deployed interactive hacks like Snake Game. Led team lessons as Scrum Master.
                    <br><a href='https://adyashipekar.github.io/code-constellations/hacks' target='_blank'>View Hacks</a>"""
    },
    {
      "title": "Sprint 3: Gamify Quest (Quest of Spook)",
      "date": "Late Trimester",
      "details": """Created the main character movement, collision system, and pantry logic for our 6-level mansion game.
                    <br><a href='https://github.com/users/AdyaShipekar/projects/4' target='_blank'>Kanban Board</a>"""
    },
    {
      "title": "Night @ the Museum (N@tM)",
      "date": "End of Trimester",
      "details": """Presented the Quest of Spook live. Game ran flawlessly, got great feedback from audience!
                    <br><a href='https://pages.opencodingsociety.com/gamify/mansionGame' target='_blank'>MicroBlog</a>"""
    },
    {
      "title": "Future Learning Goals",
      "date": "Looking Ahead",
      "details": """Continue improving Quest of Spook, learn backend/frontend development, and explore advanced data structures."""
    }
]

# --- HTML + JS Template ---
HTML(f"""
<div id="{container_id}" style="font-family:'Poppins',sans-serif; margin:40px auto; max-width:850px;">
  <h2 style="text-align:center;font-size:2em;margin-bottom:20px;">CSSE Journey Timeline</h2>
  <div class="timeline"></div>
</div>

<style>
  #{container_id} .timeline {{
    position: relative;
    margin: 40px 0;
    padding-left: 30px;
    border-left: 3px solid #4f46e5;
  }}
  #{container_id} .timeline-event {{
    position: relative;
    margin-bottom: 30px;
    cursor: pointer;
    transition: transform 0.2s ease, opacity 0.6s ease, transform 0.6s ease;
    opacity: 0;
    transform: translateY(40px);
  }}
  #{container_id} .timeline-event.visible {{
    opacity: 1;
    transform: translateY(0);
  }}
  #{container_id} .timeline-dot {{
    position: absolute;
    left: -10px;
    top: 6px;
    width: 16px;
    height: 16px;
    background: #4f46e5;
    border-radius: 50%;
  }}
  #{container_id} .timeline-content {{
    background: #f9fafb;
    padding: 12px 16px;
    border-radius: 8px;
    box-shadow: 0 3px 8px rgba(0,0,0,0.08);
  }}
  #{container_id} .timeline-title {{
    font-weight: 600;
    color: #111827;
    margin-bottom: 4px;
  }}
  #{container_id} .timeline-date {{
    font-size: 0.9em;
    color: #6b7280;
  }}
  #{container_id} .timeline-details {{
    display: none;
    font-size: 0.95em;
    color: #374151;
    margin-top: 8px;
    border-left: 3px solid #e5e7eb;
    padding-left: 10px;
  }}
  #{container_id} .timeline-event.active .timeline-details {{
    display: block;
  }}
</style>

<script>
(function() {{
  const data = {json.dumps(timeline_data)};
  const container = document.querySelector("#{container_id} .timeline");

  // Create events
  data.forEach(item => {{
    const div = document.createElement("div");
    div.classList.add("timeline-event");
    div.innerHTML = `
      <div class="timeline-dot"></div>
      <div class="timeline-content">
        <div class="timeline-title">${{item.title}}</div>
        <div class="timeline-date">${{item.date}}</div>
        <div class="timeline-details">${{item.details}}</div>
      </div>
    `;
    div.addEventListener("click", () => div.classList.toggle("active"));
    container.appendChild(div);
  }});

  // Scroll animation with IntersectionObserver
  const observer = new IntersectionObserver((entries) => {{
    entries.forEach(entry => {{
      if (entry.isIntersecting) {{
        entry.target.classList.add("visible");
        observer.unobserve(entry.target);
      }}
    }});
  }}, {{ threshold: 0.1 }});

  document.querySelectorAll("#{container_id} .timeline-event").forEach(el => observer.observe(el));
}})();
</script>
""")
```
